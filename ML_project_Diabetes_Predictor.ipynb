Importing Libraries

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Loading and Analysing Data
PIMA Indian Diabetes dataset

```python
df = pd.read_csv('/content/drive/MyDrive/ML Project Datasets/diabetes.csv')
```

0 represents non diabetic
1 represents diabetic

```python
df.groupby('Outcome').mean()
```

seperating features and targert

```python
X = df.drop(columns = 'Outcome', axis = 1)
y = df['Outcome']
```

```python
X
```

```python
y
```

Feature Scaling

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
X = scaler.transform(X)
```

```python
X[0]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify = y, random_state=1)
```

```python
X_test.shape
```

Model Training

```python
model = svm.SVC(kernel = 'linear')
```

```python
model.fit(X_train, y_train)
```

model.predict()

Evaluating the model

```python
# Accuracy Score on training data
train_prediction = model.predict(X_train)
train_accuracy = accuracy_score(train_prediction, y_train)
train_accuracy
```

```python
# accuracy on test data
test_prediction = model.predict(X_test)
test_accuracy = accuracy_score(test_prediction, y_test)
test_accuracy
```

Incorporating the trained model into a diabetes predictor function

```python
def diabetes_predictor(data):
  data = np.asarray(data)
  data = data.reshape(1, -1)
  data = scaler.transform(data)
  result = model.predict(data)
  if result < 0.5:
    print("You're alright")
  else:
    print("I'm sorry you're diabetic")
  return result

for i in range(10,20):
  print(diabetes_predictor(df.drop(columns = 'Outcome',axis = 1).iloc[i]))
```
